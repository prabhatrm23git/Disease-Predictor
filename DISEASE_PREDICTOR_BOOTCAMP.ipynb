**DAY 2** : 23/08/2025 MODEL TRAINING

```python
cat_cols = df.select_dtypes(include= 'object').columns.tolist()


if 'num' in cat_cols:
  cat_cols.remove('num')
```

```python
x = df.drop('num' , axis=1)
y = (df['num'] > 0).astype(int)
```

```python
x= pd.get_dummies(x, columns = cat_cols, dummy_na=False, drop_first=True)
print("Final feature columns : ", x.columns)
```

**DAY 3** : 24/08/2025 Advanced Models & Feature Hype

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

new mean value = (x-mean)/ std deviation

```python
#scale numeric festures
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train_scaled = scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression #classification
# from sklearn.metrics import accuracy_score, classification_report,
# confusion_matrix


#modelling
lr_model = LogisticRegression()
lr_model.fit(X_train_scaled, y_train) #train step | has correct values
```

```python
from sklearn.metrics import accuracy_score, classification_report #, confusion_matrix
```

```python
#model evaluation

y_pred_lr = lr_model.predict(X_test_scaled)
print("Logistic Regressiion Accuracy: ", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
#print(confusion_matrix(y_test, y_pred_lr))
```

DAY 4: Random Forest and Feature Importance

```python
#avoid overfitting

from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Logistic Regression)')

plt.show()
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42) #100 trees in whole forest of data set
rf_model.fit(X_train_scaled, y_train)
y_pred_rf = rf_model.predict(X_test_scaled)
print("Random Forest Accuracy: ", accuracy_score(y_test, y_pred_rf))
# print(classification_report(y_test, y_pred_rf))
# #
```

Features important

```python
feat_imp = pd.Series(rf_model.feature_importances_, index=x.columns)
feat_imp.nlargest(10).plot(kind='barh') #top 10 features
plt.title("Random Forest Feature Importance")
plt.show()
```

Save the model

```python
import joblib
joblib.dump(rf_model, 'heart_rf_model.pkl')
joblib.dump(scaler, 'heart_scaler.pkl')
```

```python
import pandas as pd
import joblib

user_df = pd.read_csv('heart_dataset.csv')
```

```python
#getting columns list from training dataframe
numeric_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
bool_cols = df.select_dtypes(include='bool').columns.tolist()

# Dropping columns which are extra in user_df tha  required to avoid error
numeric_cols = [col for col in numeric_cols if col in user_df.columns]
cat_cols = [col for col in cat_cols if col in user_df.columns]
bool_cols = [col for col in bool_cols if col in user_df.columns]

# fill the missing numeric column & cat column

user_df[numeric_cols] = user_df[numeric_cols].fillna(user_df[numeric_cols].mean())

for col in cat_cols:
  user_df[col] = user_df[col].fillna('unknown')

for col in bool_cols:
  user_df[col] = user_df[col].astype( int)

```

```python
#one hot encoding cat columns
user_df_encoded = pd.get_dummies(user_df, columns=cat_cols)

#allign columns
user_df_encoded = user_df_encoded.reindex(columns=x.columns, fill_value=0)

# Scale data
scaler = joblib.load('heart_scaler.pkl')
user_df_scaled = scaler.transform(user_df_encoded)

#prediction
model = joblib.load('heart_rf_model.pkl')
preds = model.predict(user_df_scaled)
user_df['Heart_Disease_Prediction'] = preds

print(user_df)

```
